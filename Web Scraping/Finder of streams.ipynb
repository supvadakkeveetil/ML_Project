```python
import pandas as pd
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
import random
import time

min_sleep_time = 4
max_sleep_time = 7

# Remember to change to your csv path
df = pd.read_csv('processed_data3 - Fixed.csv')
```

```python
# Define the URL column name and target element selector
base_url = "https://open.spotify.com/track/" 

# Create an empty list to store the extracted data
extracted_data = []
```

```python
driver = webdriver.Chrome()

for url in df['track_id']:
    broken_link_check = 0
    while True:
        try:
            full_url = f"{base_url}{url}"  # Use f-string for string formatting

            # Sleep for a random duration
            sleep_time = random.uniform(min_sleep_time, max_sleep_time)
            driver.get(full_url)
            time.sleep(sleep_time)

            # Find the element with the specified class name
            element_class = driver.find_elements(By.CLASS_NAME, "RANLXG3qKB61Bh33I0r2")

            # Check if any elements were found
            if element_class:
                target_data = element_class[-1].text
                print(f"Data Found: {target_data}")
                extracted_data.append(target_data)
                break  # Exit the loop after successful extraction

        except Exception as e:  # Catch specific exceptions for better error handling
            print(f"Error encountered: {e}")
            broken_link_check += 1

        # Exit loop after retry limit (consider increasing limit if needed)
        if broken_link_check == 20:
            print(f"Error Collecting Data for song {url}. Reached retry limit.")
            extracted_data.append('Not Found')
            break

# Close the browser session
driver.quit()
```

```python
# Add the extracted data as a new column in the DataFrame
df['Stream count'] = extracted_data

df.to_csv('processed_data5.csv', index=False)
```
